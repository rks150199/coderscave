```python
# Handling Null values
```

```python
data=data[['eventid', 'iyear', 'imonth', 'iday', 'country', 'region','success', 'suicide', 'targtype1',]]
data.head()
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
data.isnull().any()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
data.head()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='targtype1', data=data, palette="rocket")
plt.xlabel('target type')
plt.ylabel('No. of attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
# Most number of type of attacks was done by Dominican Republic
```

```python
# Region Wise Attacks
```

```python
ra = data["region"].value_counts().head(10)
print(ra)
```

```python
# Created the bar plot using a gradient color palette
```

```python
sns.barplot(x=ra.index, y=ra.values, palette="viridis")

plt.xlabel('Region')
plt.ylabel('Number of attacks')
plt.title('Attacks Region Wise')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
