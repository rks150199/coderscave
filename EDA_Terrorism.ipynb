```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("globalterrorismdb_0718dist.csv",encoding='ISO-8859-1')
data.head()
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
df=df[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df2=df.fillna(0)
df2.isnull().sum()
```

```python
df2.head()
```

```python
df2.describe()
```

Observation:
- terrorist activities ranging from the year: 1970 to 2017
- Max no. of people killed in an event were: 1570
- Max no. of people wounded in an event were: 8191

```python
attack=df2["iyear"].value_counts(dropna=False).sort_index()
attack.head()
```

```python
attack.plot(kind="bar",color="cornflowerblue",figsize=(15,6),fontsize=13)
plt.title("Timeline of Attacks",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

Max no. of attacks 16903 in 2014

Min no. of attacks 471 in 1971

```python
yk=df2[["iyear","nkill"]].groupby("iyear").sum()
yk.head()
```

```python
yw=df2[["iyear","nwound"]].groupby("iyear").sum()
yw.head()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)

#Killed
yk.plot(kind="bar",color="cornflowerblue",figsize=(15,15),ax=ax0)
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("Years")
ax0.set_ylabel("Number of People Killed") 
#Wounded
yw.plot(kind="bar",color="cornflowerblue",figsize=(15,15),ax=ax1)
ax1.set_title("People Wounded in each Year")
ax1.set_xlabel("Years")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```

```python
reg=pd.crosstab(df2.iyear,df2.region_txt)
reg.head()
```

```python
reg.plot(kind="area", stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```
